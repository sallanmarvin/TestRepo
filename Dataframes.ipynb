```python
# access data of all columns where mouse data is
# more than 200.

import pandas as pd
data=[{'Mouse':'150200300400'},{'Keyboard': '180200190300'}, {'Scanner':'200280330450'}]
df=pd.DataFrame(data)

# Convert the 'Mouse' column to numeric, handling potential errors
df['Mouse'] = pd.to_numeric(df['Mouse'], errors='coerce')

# Filter rows where 'Mouse' data is greater than 200
filtered_df = df[df['Mouse'] > 200]

# Print or use the filtered DataFrame
filtered_df
```
