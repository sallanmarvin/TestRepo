**Fillna()**

Pandas has .fillna() method which accepts a value which we want to replace in place of NaN values. We just calculated mean of the column and passed it as an input argument to fillna() method
DataFrame.fillna(value=None, method=None, axis=None, inplace=False, limit=None, downcast=None)


```python
df = df.fillna(0)
df["col_name"] = df["col_name"].fillna("")
df[["col1", "col2"]] = df[["col1", "col2"]].fillna("")
```

```python
df['SAFETY_SCORE'].fillna(df['SAFETY_SCORE'].mean(), inplace=True)
df['Environment_Score'].fillna(df['Environment_Score'].median(), inplace=True)
df['Instruction_Score'].fillna(df['Instruction_Score'].mean(), inplace=True)
```

```python
df['Family_Involvement_Icon'].fillna(df['Family_Involvement_Icon'].mode()[0], inplace=True)
df['Environment_Icon'].fillna(df['Environment_Icon'].mode()[0], inplace=True)
```

```python
#Examples:
#Fill with a constant value:

df.fillna(0, inplace=True)  # Replaces NaN with 0
#Fill with column mean:

df['Age'].fillna(df['Age'].mean(), inplace=True)  # Replaces NaN in 'Age' column with its mean
#Fill forward (previous value):

df.fillna(method='ffill', inplace=True)  # Forward fills NaN with previous row's value

df.fillna(method='bfill', inplace=True)  # Backward fills NaN with next row's value
```

#**DATA NORMALIZATION**

```python
import pandas as pd
import numpy as np

# Explanation of Data Normalization
"""
Data normalization is the process of scaling data so that it falls within a specific range or distribution.
It is essential in machine learning to ensure that features contribute equally to the model.
There are three common techniques for normalization:

1. **Min-Max Scaling** (Normalization)
   - Scales the data to a fixed range, usually [0,1] or [-1,1].
   - Formula: X_scaled = (X - X_min) / (X_max - X_min)
   - Useful when we need to preserve relationships in data and when data is bounded.
   - Commonly used in algorithms like k-NN, neural networks, etc.

   **Pros:**
   - Preserves the original distribution of data.
   - Keeps all values within a fixed range, preventing large feature dominance.

   **Cons:**
   - Sensitive to outliers since min and max values are used for scaling.

2. **Z-score Standardization** (Standardization)
   - Transforms data to have a mean (μ) of 0 and a standard deviation (σ) of 1.
   - Formula: X_standardized = (X - μ) / σ
   - Useful when data follows a Gaussian (Normal) distribution.
   - Commonly used in algorithms like logistic regression, SVM, and PCA.

   **Pros:**
   - Less sensitive to outliers compared to Min-Max scaling.
   - Useful when features have different scales and units.

   **Cons:**
   - Does not bound data within a specific range, which might not be ideal for some ML models.

3. **Simple Feature Scaling**
   - Scales the data by dividing each value by the maximum absolute value of the feature.
   - Formula: X_scaled = X / X_max
   - Ensures all values are in the range [0,1].
   - Suitable for cases where only positive values are expected.

   **Pros:**
   - Easy to implement and computationally efficient.
   - Works well when all values are positive.

   **Cons:**
   - Does not handle negative values well.
   - Sensitive to outliers as it depends on the max value.
"""

# Creating a sample DataFrame with different feature ranges
data = {
    'Feature1': [10, 20, 30, 40, 50],  # Small range
    'Feature2': [100, 200, 300, 400, 500],  # Larger range
    'Feature3': [5, 500, 1000, 1500, 2000]  # Highly varied range
}
df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

# Min-Max Scaling without using Scaler
df_min_max = (df - df.min()) / (df.max() - df.min())

print("\nMin-Max Scaled DataFrame (Values between 0 and 1):")
print(df_min_max)

# Z-score Standardization without using Scaler
df_zscore = (df - df.mean()) / df.std()

print("\nZ-score Standardized DataFrame (Mean = 0, Std Dev = 1):")
print(df_zscore)

# Simple Feature Scaling without using Scaler
df_simple_scaled = df / df.max()

print("\nSimple Feature Scaled DataFrame (Values between 0 and 1):")
print(df_simple_scaled)

"""
Observations:
1. **Min-Max Scaling** ensures all values lie within a specific range, making it useful for bounded data.
2. **Z-score Standardization** centers data around zero but does not bound it, which is useful for normally distributed data.
3. **Simple Feature Scaling** provides an easy way to scale features to a range of [0,1] but is sensitive to outliers.
4. **Outliers** significantly impact Min-Max scaling and Simple Feature Scaling, while Z-score handles them better.

When to Use Which:
- Use **Min-Max Scaling** when you need data within a fixed range (e.g., image processing, neural networks).
- Use **Z-score Standardization** when data needs to be normally distributed or contains outliers (e.g., PCA, linear regression).
- Use **Simple Feature Scaling** when you need a quick and computationally efficient method but without handling outliers.
"""
```
